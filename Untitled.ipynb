```python
def class_likelihood_params(class_training_data):
    
    class_likelihood_params_dict = dict()


    for cols in class_training_data.columns[0:545]:
  
        empty_dict = dict()

        if class_training_data[cols].dtype == "int64":
    

            for value in data[cols].unique():
    

                if value not in class_training_data[cols].unique():
            
                    empty_dict[value] = 1/(len(class_training_data)+len(data[cols].unique()))

                else:
            
                    empty_dict[value] = len(class_training_data[class_training_data[cols] == value])/len(class_training_data)

        else: 

            empty_dict["class_normal_mean"] = np.mean(class_training_data[cols])
            empty_dict["class_normal_std"] = np.std(class_training_data[cols]) 

        class_likelihood_params_dict[cols] = empty_dict

    return class_likelihood_params_dict
```

```python
influenza_likelihood_params_dict = class_likelihood_params(training_data[training_data[training_data.columns[545]] == 1])
```

```python
influenza_likelihood_params_dict
```

```python
noinfluenza_likelihood_params_dict = class_likelihood_params(training_data[training_data[training_data.columns[545]] == 0])
```

```python
noinfluenza_likelihood_params_dict
```

```python
import scipy.stats as s
```

```python
def class_conditioned_likelihood_per_row(params_dict, cv_data_row):
    
    p_features_on_class = 1

    for cols in cv_data.columns[0:545]:

        if cv_data[cols].dtype == "int64":

            p_features_on_class = p_features_on_class * params_dict[cols][cv_data_row[cols]]

        else:

            p_features_on_class = p_features_on_class * s.norm.pdf(cv_data_row[cols],
                                                               params_dict[cols]["class_normal_mean"],
                                                               params_dict[cols]["class_normal_std"])
        
    return p_features_on_class
```

```python
influenza_prior = len(training_data[training_data['Target'] == 1])/len(training_data)
noinfluenza_prior = len(training_data[training_data['Target'] == 0])/len(training_data)
```

```python
def fetch_predictions(data):
    
    predictions = list()

    for i in range(0,data.shape[0]):
        
        influenza_posterior_i = class_conditioned_likelihood_per_row(influenza_likelihood_params_dict,data.iloc[i,0:545]) * influenza_prior

        noinfluenza_posterior_i = class_conditioned_likelihood_per_row(noinfluenza_likelihood_params_dict,data.iloc[i,0:545]) * noinfluenza_prior

        predictions.append(influenza_posterior_i > noinfluenza_posterior_i)

    return np.array(predictions)
```

```python
cv_predictions = fetch_predictions(cv_data)
```

```python
cv_gt_labels = np.array(cv_data.iloc[:,545])
```

```python
def fetch_performance_metrics(predictions,gt_labels):

  accuracy = np.count_nonzero(predictions == gt_labels)/gt_labels.shape[0]

  return {"Accuracy":accuracy}
```

```python
cv_performance_metrics = fetch_performance_metrics(cv_predictions,cv_gt_labels)
```
